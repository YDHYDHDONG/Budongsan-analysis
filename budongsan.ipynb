- 월세 0인 경우 제외

```python
data.query('월세 <= 0')
```

```python
# 따옴표로 묶었지만 0 은 문자열이 아님 만약 데이터값이 문자열이라도 숫자로 변환함
data = data.query('월세 > 0') # query는 Dataframe 필터링 한다는 느낌으로 알면됨
```

```python
data['보증금'].unique()
```

```python
data = data.query('~보증금.str.contains("억")')
```

```python
data['보증금'] = data['보증금'].str.replace(',','').astype(int)
data['보증금'].unique()
```

- 물건층과 전체층 분리 후 1층, 꼭대기층 유뮤 컬럼 만들기
  - 지하의 경우 마이너스로 표시

```python
data[['물건층', '전체층']] = data['층수(물건층/전체층)'].str.split('/', expand=True)
```

```python
data.head()
```

```python
data['물건층'].unique()
```

```python
def floor_info(target, total):
  try:
    if target in ['B1','B2']: #지하이면
      return '비선호'
    elif int(target) == 1 or int(target)/int(total)== 1: #1층이거나 꼭대기층이면
      return '비선호'
    else:
      return '선호'
  except ValueError:
    return '선호'
```

```python
data['비선호층여부'] = data.apply(lambda x: floor_info(x['물건층'],x['전체층']), axis=1)
```

```python
data.query('물건층 == "B1"')
```

- 제외 조건에 해당하는 매물 제외
  - 원하는 조건
    - 보증금 3000만원 이하
    - 월세는 저렴할수록 좋음
    - 지하, 반지하, 꼭대기층은 선호하지 않음
    - 전용면적이 클수록 좋음
    - 북향 비선호
    - 연식이 오래되지 않을수록 좋음
    - 지하철 역에서 가까울수록 좋음

```python
#보증금 3000만원 이하
#지하, 반지하, 꼭대기층은 비선호
#북향 비선호
data_filtered = data.query('300<=보증금<=3000 and 비선호층여부 == "선호" and 전체층 != "1" and  ~방향.str.contains("북")')
```

```python
data_filtered
```

```python
data_filtered[['tag1','tag2','tag3','tag4']] = data_filtered['태그'].str.replace("\'|\[|\]","").str.split(', ', expand=True)
```

```python
data_filtered.head()
```

```python
data_filtered = data_filtered.query('tag1.str.contains("년")') #연식 정보가 있는 데이텀나 필터링
```

```python
data_filtered['연식'] = [int(i[0]) for i in data_filtered['tag1'].str.split('년')]
data_filtered
```
