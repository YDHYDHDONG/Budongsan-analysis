```python
# 따옴표로 묶었지만 0 은 문자열이 아님 만약 데이터값이 문자열이라도 숫자로 변환함
data = data.query('월세 > 0') # query는 Dataframe 필터링 한다는 느낌으로 알면됨
```

```python
data['보증금'].unique()
```

```python
data = data.query('~보증금.str.contains("억")')
```

```python
data['보증금'] = data['보증금'].str.replace(',','').astype(int)
data['보증금'].unique()
```

- 물건층과 전체층 분리 후 1층, 꼭대기층 유뮤 컬럼 만들기
  - 지하의 경우 마이너스로 표시

```python
data[['물건층', '전체층']] = data['층수(물건층/전체층)'].str.split('/', expand=True)
```

```python
data.head()
```

```python
data['물건층'].unique()
```

```python
def floor_info(target, total):
  try:
    if target in ['B1','B2']: #지하이면
      return '비선호'
    elif int(target) == 1 or int(target)/int(total)== 1: #1층이거나 꼭대기층이면
      return '비선호'
    else:
      return '선호'
  except ValueError:
    return '선호'
```

```python
data['비선호층여부'] = data.apply(lambda x: floor_info(x['물건층'],x['전체층']), axis=1)
```
